```python
# %load_ext autoreload
# %autoreload 2

from morph import inflect, get_tag, morph

def swap_person(word: str) -> str:
    pronoun = [
        ('я', 'ты'),
        ('меня', 'тебя'),
        ('мне', 'тебе'),
        ('мной', 'тобой'),
        ('мои', 'твои'),
        ('моих', 'твоих'),
        ('моим', 'твоим'),
        ('моими', 'твоими'),
        ('моих', 'твоих'),
        ('наш', 'ваш'),
        ('нашего', 'вашего'),
        ('нашему', 'вашему'),
        ('нашими', 'вашими'),
        ('наших', 'ваших'),
    ]

    for p in pronoun:
        if p[0] == word:
            return p[1]

        if p[1] == word:
            return p[0]

    for parsed in morph.parse(word):
        if 'impr' in parsed.tag:
            inflected = parsed.inflect({ 'indc', 'futr' }) or parsed.inflect({'indc' })
            if inflected:
                return inflected.word

        if '2per' in parsed.tag:
            inflected = parsed.inflect({ '1per' })
            if inflected:
                return inflected.word

        if '1per' in parsed.tag:
            inflected = parsed.inflect({ '2per' })
            if inflected:
                return inflected.word

    return word

def swap_person_in_text(text: str) -> str:
    return ' '.join([swap_person(w) for w in text.split(' ')])
```
